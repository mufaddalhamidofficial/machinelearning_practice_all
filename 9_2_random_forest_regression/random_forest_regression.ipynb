## Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the data

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training Random Forest Reg model

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 10)

regressor.fit(X, y)
```

## Visualising Random Forest Reg results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title("Decsion Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualising Poly Reg with higher resolution and smoother curve

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title("Decision Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Score of Linear and Poly

```python
print(regressor.score(X, y))
```

## Predicting a new result with both models

```python
val = float(input("Enter position level"))
print(regressor.predict([[val]])[0])
```
