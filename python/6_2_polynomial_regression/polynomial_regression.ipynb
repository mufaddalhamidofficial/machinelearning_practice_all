## Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the data

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training Linear Reg model

```python
from sklearn.linear_model import LinearRegression

linearReg = LinearRegression()

linearReg.fit(X, y)
```

```python
linearReg.predict(X)
```

## Training Polynomial Reg model

```python
from sklearn.preprocessing import PolynomialFeatures

polynomialReg = PolynomialFeatures(degree = 10)

X_poly = polynomialReg.fit_transform(X, y)
```

```python
linearPolyReg = LinearRegression()
linearPolyReg.fit(X_poly, y)
```

## Visualising Linear Reg results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, linearReg.predict(X), color = 'blue')
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualising Polynomial Regression Results


```python
plt.scatter(X, y, color = 'red')
plt.plot(X, linearPolyReg.predict(X_poly), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualising Poly Reg with higher resolution and smoother curve

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, linearPolyReg.predict(polynomialReg.fit_transform(X_grid)), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Score of Linear and Poly

```python
print(linearReg.score(X, y))
print(polynomialReg.score(X, y))
```

## Predicting a new result with both models

```python
val = int(input("Enter position level"))
print(linearReg.predict([[val]])[0])
print(linearPolyReg.predict(polynomialReg.fit_transform([[val]]))[0])
```
